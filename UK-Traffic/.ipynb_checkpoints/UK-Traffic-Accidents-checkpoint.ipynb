```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from kaggle.api.kaggle_api_extended import KaggleApi


api = KaggleApi()
api.authenticate()
```

```python
accident_data_path = 'bluehorseshoe/uk-2016-road-safety-data'
accident_data_files = api.dataset_list_files(accident_data_path).files
print(accident_data_files)
```

```python
# download everything and load data from the relevant files
for file in accident_data_files:
    api.dataset_download_file(accident_data_path, file.name)

district_geo_file = "./Local_Authority_Districts_Dec_2016.geojson" 
district_geo = gpd.read_file(district_geo_file)
district_geo.head()
```

```python
traffic_data_file = "./dftRoadSafety_Accidents_2016.csv.zip" 
traffic_data = pd.read_csv(traffic_data_file)
traffic_data.head()

#TODO: add parameters specifying the data type for the mixed data columns to read_csv
# https://stackoverflow.com/questions/24251219/pandas-read-csv-low-memory-and-dtype-options
```

```python
# convert the traffic data to a GeoDataFrame
traffic_data_gdf = gpd.GeoDataFrame(
    traffic_data, geometry=gpd.points_from_xy(traffic_data.Longitude, traffic_data.Latitude))
traffic_data_gdf.head()
```

```python
# map the data
ax = district_geo.plot(figsize=(10,10), color='none', edgecolor='gainsboro')
traffic_data_gdf.plot(color="maroon", markersize=2, ax=ax)
```
