```python
import pandas
import matplotlib.pyplot
from kaggle.api.kaggle_api_extended import KaggleApi
import folium


api = KaggleApi()
api.authenticate()
```

```python
# list the files available from the Kaggle data set
accident_data_path = 'bluehorseshoe/uk-2016-road-safety-data'
accident_data_files = api.dataset_list_files(accident_data_path).files
print(accident_data_files)
```

```python
# make a Pandas dataframe of traffic data file
api.dataset_download_file(accident_data_path, "dftRoadSafety_Accidents_2016.csv")
traffic_data_filepath = "./dftRoadSafety_Accidents_2016.csv.zip"
traffic_data = pandas.read_csv(traffic_data_filepath)

# remove any rows missing coordinates
traffic_data = traffic_data.dropna(subset=['Longitude', 'Latitude'])

traffic_data.head()
```

```python
# create a map with a pointer for each accident location
locations = traffic_data[['Latitude', 'Longitude']]
locationlist = locations.values.tolist()

traffic_map = folium.Map(location=[51.5074, 0.1278], zoom_start=10)

# test mapping the first 100 locations
for location in locationlist[0:99]:
    folium.Marker(location).add_to(traffic_map)

traffic_map
```
